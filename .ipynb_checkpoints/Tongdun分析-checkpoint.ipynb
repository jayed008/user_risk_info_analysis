```python
import matplotlib.pyplot as plt
#查看用户借款平台的分布
def find_platform_freq(a,plt_str):
    result = db.ResultS8.find({'isRepay':a})
    platform_list = []
    for r in result:
        rr = r['risk_items']
        risk_len = len(rr)
        for rrr in rr:
            if rrr['item_name'] == plt_str:
                platform = rrr['item_detail']['platform_detail']
                for p in platform:
                    p1 = re.compile(r'(.*)[:]', re.S)
                    p2 = re.findall(p1,p)[0]
                    p3 = re.compile(r'[:](.*)', re.S)
                    p4 = re.findall(p3,p)[0]
#                     print(p2,p4)
                    for i in range(int(p4)):
                        platform_list.append(p2)
    return Counter(platform_list)

#将借款平台频数分布转化为频率分布，并将Counter转化为dict
def get_01_freq_dict_plt(d0,d1):
    d0_platform_amo = 0
    d1_platform_amo = 0
    for v0 in d0.values():
        d0_platform_amo +=v0
    for v1 in d1.values():
        d1_platform_amo +=v1

    d0_platform_dict = {}
    d1_platform_dict = {}

    for pf in d0.keys():
        d0_platform_dict.update({pf:d0[pf]/d0_platform_amo})

    for pf in d1.keys():
        d1_platform_dict.update({pf:d1[pf]/d1_platform_amo})
        
    return d0_platform_dict,d1_platform_dict

#画字典的柱形图
def plot_01_dict_bar(d0,d1,title_str):
    sns.set(context='notebook', font= 'simhei',style='whitegrid')
    C0_list = []
    C0_columns = []
    C1_list = []
    C1_columns = []
    for key0,key1 in zip(d0.keys(),d1.keys()):
        C0_list.append(d0[key0])
        C0_columns.append(key0)
        C1_list.append(d1[key1])
        C1_columns.append(key1)
    C0_se =Series(C0_list,index=C0_columns)
    C1_se =Series(C1_list,index=C1_columns)

    C_df = DataFrame({"0":C0_se,"1":C1_se})
    tstr = '还款用户(1)与非还款用户(0)' + title_str + '指标占比情况'
    C_df.plot(kind="bar",alpha=0.7,figsize = (20,5),fontsize = 12, legend=True, title=tstr)

#还款用户与非还款用户platform指标占比情况('15天申请人在多个平台申请借款')
d0_15_platform = find_platform_freq(0,'15天申请人在多个平台申请借款')
d1_15_platform = find_platform_freq(1,'15天申请人在多个平台申请借款')
d0_15_platform_dict,d1_15_platform_dict = get_01_freq_dict_plt(d0_15_platform,d1_15_platform)
plot_01_dict_bar(d0_15_platform_dict,d1_15_platform_dict,'15days_platform')
```

```python
#画出身份证和手机命中风险分布
import matplotlib.pyplot as plt

#查看用户风险类型的分布
def find_risk_freq(a,type_str):
    result = db.ResultS8.find({'isRepay':a})
    fraud_type_list = []
    for r in result:
        rr = r['risk_items']
        risk_len = len(rr)

        for rrr in rr:
            if rrr['item_name'] == type_str:
                fraud_type = rrr['item_detail']['fraud_type']
    #             print(fraud_type)
    #             p1 = re.compile(r'\u3001')
    #             p2 = re.findall(p1,fraud_type)
                p3 = re.split(r'\u3001', fraud_type)
    #             print(p3)
                for p4 in p3:
                    fraud_type_list.append(p4)
    id_risk = Counter(fraud_type_list)
    return id_risk

#将频数转化为频率，并将Counter转化为字典
def get_01_freq_dict(d0,d1):
    amo0 = 0
    amo1 = 0
    for v0 in d0.values():
        amo0 +=v0
    for v1 in d1.values():
        amo1 +=v1

    d0_dict = {}
    d1_dict = {}
    for pf in d0.keys():
        d0_dict.update({pf:d0[pf]/amo0})

    for pf in d1.keys():
        d1_dict.update({pf:d1[pf]/amo1})
    
    return d0_dict,d1_dict

#画出用户风险类型分布('身份证命中中风险关注名单')
d0_id_risk = find_risk_freq(0,'身份证命中中风险关注名单')
d1_id_risk = find_risk_freq(1,'身份证命中中风险关注名单')
d0_id_risk_dict, d1_id_risk_dict = get_01_freq_dict(d0_id_risk,d1_id_risk)
plot_01_dict_bar(d0_id_risk_dict,d1_id_risk_dict,'id_risk')

#画出用户风险类型分布('手机号命中中风险关注名单')
d0_phone_risk = find_risk_freq(0,'手机号命中中风险关注名单')
d1_phone_risk = find_risk_freq(1,'手机号命中中风险关注名单')
d0_phone_risk_dict, d1_phone_risk_dict = get_01_freq_dict(d0_phone_risk,d1_phone_risk)
plot_01_dict_bar(d0_phone_risk_dict,d1_phone_risk_dict,'phone_risk')
```

```python
def find_id_linked_amo(a):
    result = db.ResultS8.find({'isRepay':a})
    amo_list = []
    for r in result:
        rr = r['risk_items']
        for rrr in rr:
            if rrr['item_name'] == '3个月内身份证关联多个申请信息':
                rrrr = rrr['item_detail']['frequency_detail_list']
                p1 =0
                p2 =0
                for r0 in rrrr:
                    if '3个月身份证关联家庭地址数' in r0['detail']:
                        p1 = re.findall(r'[：](.*)', r0['detail'])[0]
                       
                    if '3个月身份证关联手机号数' in r0['detail']:
                        p2 = re.findall(r'[：](.*)', r0['detail'])[0]
                        
                p = int(p1)+int(p2) #用户的身份证关联信息总数
                amo_list.append([int(p1),int(p2),p])
    amo_list = np.array(amo_list)
    return Counter(amo_list[:,0]),Counter(amo_list[:,1]),Counter(amo_list[:,2])

d00_id_linked_amo,d01_id_linked_amo,d02_id_linked_amo = find_id_linked_amo(0)
d10_id_linked_amo,d11_id_linked_amo,d12_id_linked_amo = find_id_linked_amo(1)


d02_id_linked_amo_dict, d12_id_linked_amo_dict = get_01_freq_dict(d02_id_linked_amo,d12_id_linked_amo)
plot_01_dict_bar(d02_id_linked_amo_dict,d12_id_linked_amo_dict,'id_linked_amo_2')
```
